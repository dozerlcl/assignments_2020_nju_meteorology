```python
# from sympy import symbols, cos, sin
# from sympy.plotting import plot_parametric
import sympy as sym
import numpy as np
r  = sym.symbols('r')
dh = sym.symbols('deltah')
# x     = sym.symbols('x')
# Q0    = 340.
# xs    = 0.95
# A=211.1
# B=1.55
# Tsnow = -10.
# sx    = 1 - 0.241*(3*x**2-1)
held_H = sym.sqrt((5/3 * r * dh))
modidifed_H = 15/8. * r * dh
```

```python
held_H
```

```python
modidifed_H
```

```python
# from sympy import plot_implicit, cos, sin, symbols, Eq, And

import seaborn as sns
sns.set()
sns.set_style("whitegrid", {'grid.linestyle': '--'})

y    = sym.symbols('held')
ynew = sym.symbols('modified')
y    = held_H.subs({dh:1/3})
ynew = modidifed_H.subs({dh:1/3})
p1 = sym.plot(y,(r,0,1.7),adaptive=False,line_color='r',show=False)
p2 = sym.plot(ynew,(r,0,1.7),adaptive=False,line_color='k',show=False)
# p2 = sym.plot_implicit(sym.Eq(y,Qx_new),(y,0.8,1.2),(x,0,1),adaptive=False,line_color='k')
# p2 = plot(Qx_new, (x,0,1), show=False)
p1.extend(p2)
p1.show()
```
