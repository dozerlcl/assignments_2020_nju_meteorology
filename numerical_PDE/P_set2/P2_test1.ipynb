P1.1. 实际上需要做的是求方程的数值解
下面简单罗列了几种方程数值解的求法

```python
import numpy as np
import sympy as sym
from sympy import sin,cos
from scipy.optimize import fsolve, root
import matplotlib.pyplot as plt
```

```python
# 数学表达式的输入
x = sym.symbols('x') # 对应 k * delta
y = sym.symbols('y') # 对应 k'* delta
f1 = y**2+2*sym.cos(x)-2 # 对应1的方程
f2 = -10/12*y**2 - 1/12*y**2*2*sym.cos(y) - 2*sym.cos(x) + 2 #对应2问的方程
```

下面开始求解
## 1.1

```python
xx = np.linspace(0,np.pi,20)
yy_f1 = np.empty(xx.shape)
yy_f2 = np.empty(xx.shape)
for i,subs_xx in enumerate(xx):
    yy_f1[i] = sym.nsolve(f1.subs(x,subs_xx),y,0) #1.1的数值解
    yy_f2[i] = sym.nsolve(f2.subs(x,subs_xx),y,0) #1.2的数值解
```

```python
fig,ax = plt.subplots(1,figsize=(6,6))
# custom_xlim = (0, np.p)
# custom_ylim = (0, 5.5)
# Setting the values for all axes.

# plt.setp(ax, xlim=custom_xlim, ylim=custom_ylim)
ax.plot(xx**2, yy_f1**2,'-*r', label='second-order central')
ax.plot(xx**2, yy_f2**2,'-xg', label='fourth-order')
ax.plot(xx**2, xx**2,'--xb', label='correct')
# ax.set_xscale("log", basex=10)
# ax.set_yscale("log", basey=10)
ax.legend(loc='upper left')
plt.xlabel('(delta*k)^2')
plt.ylabel('(delta*k\')^2')
#     ax[nn].legend(')
```
